```python
import pandas as pd

# Assuming your dataframe is named df

# Filter the DataFrame to include only the necessary columns
filtered_df = df[["INDUSTRY", "PROFILE", "PLATFORM", "ENGAGEMENT RATE/ PROFILE"]]

# Create separate dataframes for each industry
fashion_df = filtered_df[filtered_df["INDUSTRY"] == "Fashion"]
tech_df = filtered_df[filtered_df["INDUSTRY"] == "Tech"]
fmcg_df = filtered_df[filtered_df["INDUSTRY"] == "FMCG"]
travel_df = filtered_df[filtered_df["INDUSTRY"] == "Travel"]
```

```python
## TRAVEL ANALYSIS BEGINS#######
# Sort travel_df by the PROFILE column
sorted_travel_df = travel_df.sort_values(by="PROFILE")

# Display the sorted dataframe
# sorted_travel_df
# Drop rows with NaN values in the 'ENGAGEMENT RATE/ PROFILE' column
sorted_travel_df.dropna(subset=["ENGAGEMENT RATE/ PROFILE"], inplace=True)

# Display the dataframe after dropping NaN values
sorted_travel_df
# Check the data type of the 'ENGAGEMENT' column
engagement_dtype = sorted_travel_df["ENGAGEMENT RATE/ PROFILE"].dtype
print("Data type of the 'ENGAGEMENT RATE/ PROFILE' column:", engagement_dtype)
# Remove percent sign from 'ENGAGEMENT RATE/ PROFILE' column and convert to numeric
sorted_travel_df["ENGAGEMENT RATE/ PROFILE"] = (
    sorted_travel_df["ENGAGEMENT RATE/ PROFILE"].str.rstrip("%").astype(float)
)

# Display the dataframe after conversion
sorted_travel_df.head()
```

```python
# Group by PROFILE and find the index of the row with the maximum ENGAGEMENT RATE/ PROFILE in each group
max_engagement_idx = sorted_travel_df.groupby("PROFILE")[
    "ENGAGEMENT RATE/ PROFILE"
].idxmax()

# Get the rows corresponding to the maximum engagement indices
max_engagement_rows = sorted_travel_df.loc[max_engagement_idx]

# Display the result
print(max_engagement_rows[["PROFILE", "PLATFORM", "ENGAGEMENT RATE/ PROFILE"]])
```

```python
# Count the occurrences of each platform in max_engagement_rows
platform_counts = max_engagement_rows["PLATFORM"].value_counts()

# Get the platform with the maximum count
most_common_platform = platform_counts.idxmax()

# Display the result
print(
    " Travel industry Platform with the most appearances  with the maximum engagement rate:",
    most_common_platform,
)

###########################TRAVEL DONE
```

```python
import matplotlib.pyplot as plt

# Data for Travel industry (replace this with your actual data)
travel_data = {
    "allegiant": {"instagram": 30.94},
    "carnival": {"tiktok": 16.53},
    "delta": {"instagram": 5.19},
    "expedia": {"tiktok": 6.89},
    "hilton": {"instagram": 84.61},
    "hostelworld": {"instagram": 30.19},
    "palisociety": {"tiktok": 55.25},
    "tripadvisor": {"tiktok": 27.75},
    "trivago": {"tiktok": 17.96},
    "viking": {"instagram": 8.07},
    "vrbo": {"tiktok": 13.03},
}

# Count the occurrences of each platform
platform_counts_travel = {
    platform: sum(
        1
        for profile, data in travel_data.items()
        if data.get(platform, 0) == max(data.values())
    )
    for platform in set(platform for data in travel_data.values() for platform in data)
}

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(
    platform_counts_travel.values(),
    labels=platform_counts_travel.keys(),
    autopct="%1.1f%%",
    startangle=140,
)
plt.title("Platform with Maximum Engagement Rate in Travel Industry")
plt.axis("equal")  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
